```python
# Step 1: Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import numpy as np
```

```python
# Step 2: Load the Dataset
df = pd.read_csv("brain_boost_sample_dataset_fixed.csv")
print("Dataset Loaded Successfully!")
display(df.head())
```

```python
# Step 3: Preprocess Time Columns (Convert to datetime or minutes)
# Function to count the number of coffee times
def count_coffee_times(coffee_str):
    if pd.isna(coffee_str):
        return 0
    return len(coffee_str.split(','))

df['Num_Coffee_Times'] = df['Coffee Times'].apply(count_coffee_times)

# Display the updated DataFrame with the new column
display(df[['Coffee Times', 'Num_Coffee_Times']].head())
```

```python
# Step 4: Simulate Focus Score Per Hour (0â€“24)
# Define the time_to_minutes function
def time_to_minutes(t):
    try:
        h, m = map(int, t.split(":"))
        return h * 60 + m
    except:
        return None

def simulate_focus(row):
    focus_scores = []
    # Ensure 'Wake_Minutes' and 'Sleep_Minutes' exist, handle potential NaNs
    wake = row.get("Wake_Minutes")
    sleep = row.get("Sleep_Minutes")

    # Handle potential NaN in 'Coffee Times' before splitting
    coffee_times_str = row.get("Coffee Times")
    coffee_times = coffee_times_str.split(", ") if pd.notna(coffee_times_str) else []

    # Ensure time_to_minutes function is available
    # Assuming time_to_minutes is defined in a previous cell and executed
    coffee_minutes = [time_to_minutes(t) for t in coffee_times if t]


    for hour in range(24):
        minutes = hour * 60
        base_focus = 50

        # Circadian rhythm effect
        if 9 <= hour <= 11 or 15 <= hour <= 17:
            base_focus += 20
        elif 12 <= hour <= 14:
            base_focus -= 10
        elif hour >= 21 or hour < 6:
            base_focus -= 20

        # Coffee boost
        if any(abs(minutes - c) <= 60 for c in coffee_minutes):
            base_focus += 10

        # Noise penalty
        noise_penalty = {"Low": 0, "Medium": -5, "High": -15}
        base_focus += noise_penalty.get(row["Noise Level"], 0)

        # Screen time effect (light penalty if more than 6 hours)
        if row["Screen Hours"] > 6:
            base_focus -= 5

        # Sleep quality
        sleep_bonus = {"Good": 10, "Average": 0, "Poor": -10}
        base_focus += sleep_bonus.get(row["Sleep Quality"], 0)

        focus_scores.append(max(0, min(100, base_focus)))

    return focus_scores

df["Hourly Focus"] = df.apply(simulate_focus, axis=1)

# Display the first few rows with the new column
display(df[['Hourly Focus']].head())
```

```python
# Step 5: Visualize Focus Levels of First User
focus_hours = df.loc[0, "Hourly Focus"]
hours = list(range(24))

plt.figure(figsize=(10, 5))
plt.plot(hours, focus_hours, marker='o')
plt.title("Simulated Focus Levels Throughout the Day (User 1)")
plt.xlabel("Hour of Day")
plt.ylabel("Focus Level (0-100)")
plt.xticks(hours)
plt.grid(True)
plt.show()

# Step 6: Recommend Best Focus Windows (User 1)
def get_focus_windows(focus_list, threshold=70):
    windows = []
    for hour, score in enumerate(focus_list):
        if score >= threshold:
            windows.append(hour)
    return windows

best_hours = get_focus_windows(focus_hours)
print("\nRecommended Focus Hours for User 1:", best_hours)
print("(Tip: Try deep work during these hours!)")
```

```python
# Step 7: Plot all usersâ€™ focus levels (optional: use a subset if many)
plt.figure(figsize=(12, 6))
for i in range(len(df)):
    plt.plot(range(24), df.loc[i, "Hourly Focus"], label=f"User {i+1}")
plt.title("Focus Patterns for All Users")
plt.xlabel("Hour")
plt.ylabel("Focus Level")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Step 8: Convert Hourly Focus into a DataFrame for heatmap
focus_df = pd.DataFrame(df["Hourly Focus"].to_list())
focus_df.index = [f"User {i+1}" for i in df.index]

plt.figure(figsize=(14, 6))
sns.heatmap(focus_df, cmap="YlGnBu", xticklabels=range(24), yticklabels=focus_df.index)
plt.title("Focus Level Heatmap (Users vs Hours)")
plt.xlabel("Hour of Day")
plt.ylabel("User")
plt.show()
```

```python
#Step 9: Add the best focus hours for each user
df["Best Hours"] = df["Hourly Focus"].apply(lambda x: [i for i, score in enumerate(x) if score >= 70])

# Export
df.to_excel("brain_boost_results.xlsx", index=False)
print("Results saved to brain_boost_results.xlsx")
```
